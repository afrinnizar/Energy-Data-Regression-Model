```python
x=df1.drop(columns=['Cooling load'])       #remove the cooling load from df1 and store the rest in x
```

```python
x.columns         #viewing the columns in x
```

```python
y=df1['Cooling load']       # store the cooling load columns from df1 in y(target)
```

```python
y     #calling y
```

```python
from sklearn.feature_selection import SelectKBest ,f_regression      # importing sklearn 
```

```python
sk=SelectKBest(score_func=f_regression,k='all')       #select all features using the f_regression scoring function
```

```python
x_new=sk.fit_transform(x,y)              #select all features from x based on y
```

```python
selected_columns=x.columns[sk.get_support()]     #get the names of the selected features
```

```python
selected_columns        #calling selected columns
```

```python
xselected=x[selected_columns]        #
```

```python
featurescore=pd.DataFrame({'features':x.columns,'score':sk.scores_
                          }).sort_values(by='score',ascending=False)         #create a data frame with features names and their score by score
```

```python
featurescore      #calling feature score
```

```python
from sklearn.preprocessing import StandardScaler     #importing standard scaler
```

```python
ss=StandardScaler()         #setting up the scaler 
```

```python
xscaled=ss.fit_transform(xselected)        #scaling the selected features
```

```python
xscaled                              #viewing the scaled one
```

```python
from sklearn.model_selection import train_test_split   #importing train test split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xscaled,y,test_size=0.2,random_state=42)    #split the data into train testing set
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:",x_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)      #print the shape of the train and testing test
```

```python
from sklearn.linear_model import LinearRegression     #importing the algorithm-------------*linear regression
```

```python
LR=LinearRegression()       #create a linear regression object
```

```python
LR.fit(x_train,y_train)      #training data
```

```python
ypred_LR=LR.predict(x_test)       #predicting
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score   #importing mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,ypred_LR)
mse=mean_squared_error(y_test,ypred_LR)
r2=r2_score(y_test,ypred_LR)                                        #evaluating predict values
```

```python
print(mae)
print(mse)
print(r2)                                   #displaying the predict values
```

```python
from sklearn.svm import SVR   #importing the algorithm------------*SVR*
```

```python
svr=SVR()                          #create a SVRegression object
```

```python
svr.fit(x_train,y_train)                #teaching
```

```python
ypred_svr=svr.predict(x_test)      #predicting
```

```python
mae=mean_absolute_error(y_test,ypred_svr)
mse=mean_squared_error(y_test,ypred_svr)                                  #evaluating predict values (mean_absolute_error,mean_squared_error,r2_score)
r2=r2_score(y_test,ypred_svr)
```

```python
print(mae)
print(mse)                                      #displaying the predicted values
print(r2)
```

```python
from sklearn.tree import DecisionTreeRegressor         #importing DecisionTreeRegressor
```

```python
dr=DecisionTreeRegressor()      #create a DecisionTreeRegressor object 
```

```python
dr.fit(x_train,y_train)        #Teaching 
```

```python
ypred_dr=dr.predict(x_test)                    #predicting 
```

```python
print(mae)
print(mse)                    #displaying the predicted values
print(r2)
```
